## 문제1
- 주어진 데이터셋(members.csv)의 'views' 컬럼 상위 10개 데이터를 상위 10번째 값으로 대체한 후 'age'컬럼에서 80 이상인 데이터의 'views' 컬럼 평균값 구하기


```python
# views 컬럼 - 내림차순으로 정렬 - 상위 10개 데이터
df = df.sort_values('views', ascending=False)
df.head(10)
```

```python
# 10개 중에서 10번째(최소값) 확인 / index는 0부터 시작
min_value = df['views'][:10].min()
print(min_value)
min_value = df['views'].iloc[9]
print(min_value)
min_value = df['views'].head(10).min()
print(min_value)
```

```python
# 10개 중에서 10번째 (최소값) 값 대체
df.iloc[:10, -1] = min_value
```

```python
# 대체된 데이터 확인
df.head(12)
```

```python
# age 컬럼에서 80 이상인 view
cond = df['age'] >= 80
newdata = df[cond]
print(newdata['views'].mean())
```

## 문제2
- 주어진 데이터셋(members.csv)의 앞에서부터 순서대로 80% 데이터만 활용해 'f1'컬럼 결측치를 중앙값으로 채우기 전 후의 표준편차를 구하고, 두 표준편차 차이 계산하기
(단, 표본표준편차 기준, 두 표준편차 차이는 절대값으로 계산)

```python
# 앞에서 부터 80% 데이터 뽑아오기
df = pd.read_csv('members.csv')
line = int(len(df)*0.8)
df = df.iloc[:line] #0~79
df
```

```python
## 결측치 채우기 전 f1컬럼 표준편자
std1 = df['f1'].std()
std1
```

```python
## 결측치 확인
df.isnull().sum()
```

```python
# 중앙값 확인
med = df['f1'].median()
med
```

```python
## 중앙값으로 채우기
df['f1'] = df['f1'].fillna(med)
```

```python
## 결측치 확인
df.isnull().sum()
```

```python
# 결측치 채운 후 표준편차 구하기
std2 = df['f1'].std()
std2
```

```python
# 두 표준편차의 차이
print(abs(std1 - std2))
```

## 심화학습
- 모 표준편차 vs 표본 표준 편차

```python
# pandas :  표본 표준편차 / numpy : 모 표준편차
```

```python
import numpy as np
print(df['f1'].std())
print(np.std(df['f1'], ddof=1))
print(df['f1'].std(ddof=0))
print(np.std(df['f1']))
```

## 문제3
- 주어진 데이터셋(members.csv)의 'age'컬럼의 이상치를 모두 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
# 표준편차, 평균값 구하기
df = pd.read_csv('members.csv')
std1 = df['age'].std() * 1.5
std1
```

```python
mean1 = df['age'].mean()
mean1
```

```python
# 이상치 최저, 최고 기준 구하기
lower = mean1 - std1
upper = mean1 + std1

print(lower)
print(upper)
```

```python
# 이상치를 벗어나는 값(조건) 찾기
cond1 = df['age'] < lower
cond2 = df['age'] > upper

df[cond1|cond2]
```

```python
# 이상치 age 합
print(df[cond1|cond2]['age'].sum())
```

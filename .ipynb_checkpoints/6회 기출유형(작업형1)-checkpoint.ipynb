1. 주어진 데이터는 각 소방서의 출동/도착 시간데이터이다.
출동시간과 도착시간 차이가 평균적으로 가장 오래 걸린 소방서의 시간을 분으로 변환해 출력하시오. (반올림 후 정수 출력)ㅡ

```python
import pandas as pd
df = pd.read_csv('6회 작업형1/data6-1-1.csv')
df.head(2)
```

```python
df['출동시간'] = pd.to_datetime(df['출동시간'])
df['도착시간'] = pd.to_datetime(df['도착시간'])
```

```python
df['diff'] = df['도착시간'] - df['출동시간']
```

```python
df['diff'].dt.total_seconds()/ 60 ## 일까지 계산
```

```python
df['diff'].dt.seconds/ 60  ## 일을 제외한 시간만 계산
```

```python
df['diff'] = df['diff'].dt.total_seconds()/ 60 ## 일까지 계산
```

```python
df.head()
```

```python
result = df.groupby('소방서')['diff'].mean()
```

```python
result.sort_values(ascending = False)
```

```python
print(round(result.max()))
```

2. 학교에서 교사 한 명당 맡은 학생 수가 가장 많은 학교를 찾고, 그 학교의 전체 교사 수를 구하시오. (정수 출력)

```python
import pandas as pd
df = pd.read_csv("6회 작업형1/data6-1-2.csv")
df
```

```python
df['result'] = (df['1학년'] + df['2학년'] + df['3학년'] + df['4학년'] + df['5학년'] + df['6학년']) / df['교사수'] 
```

```python
result = df.sort_values('result',ascending=False)
```

```python
result
```

```python
print(result['교사수'][7])
```

```python
import pandas as pd
df = pd.read_csv("6회 작업형1/data6-1-2.csv")
df
```

```python
# 전체학생수
#df.iolc[행, 열]
df['전체'] = df.iloc[:,2:].sum(axis=1)
```

```python
# 교사 한명당 맡은 학생 수 (전체 / 교사수)
df['전체/교사수'] = df['전체'] / df['교사수']
df.head(2)
```

```python
# 전체/교사가 가장 큰값을 가진 학교의 교사수
df['전체/교사수'].idxmax()
```

```python
print(int(df.loc[7,'교사수']))
```

3. 연도별로 총 범죄 건수(범죄유형의 총합)의 월평균 값을 구한 후 그 값이 가장 큰 연도를 찾아, 해당 연도의 총 범죄 건수의 월평균 값을 출력하시오. (반올림하여 정수로 출력)

```python
import pandas as pd
df = pd.read_csv("6회 작업형1/data6-1-3.csv")
df.head(5)
```

```python
df.info()
```

```python
df['월평균'] = df.iloc[:,1:-1].sum(axis=1)/6
```

```python
df.head(5)
```

```python
df = df.sort_values('월평균',ascending=False)
df.head(2)
```

```python
print(round(df.loc[5,'월평균']))
```

```python
import pandas as pd
df = pd.read_csv("6회 작업형1/data6-1-3.csv")
df.head(5)
```

```python
# 총 범죄 건수
df['총범죄'] = df.iloc[:,1:-1].sum(axis=1)
df.head()
```

```python
# 연도
df['년도'] = df['날짜'].str[:4]
df.head()
```

```python
# 연도별 그룹핑 & 월평균
result = df.groupby('년도')['총범죄'].sum() / 12
```

```python
# 가장 큰 값
print(round(result.max()))
```

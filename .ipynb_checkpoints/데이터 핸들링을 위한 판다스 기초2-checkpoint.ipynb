# 인덱싱 / 슬라이싱

```python
# 학습 전 반드시 실행: csv파일 (data.csv) 생성
import pandas as pd
data = {
    "메뉴":['아메리카노','카페라떼','카페모카', '바닐라라떼', '녹차', '초코라떼', '바닐라콜드브루'],
    "가격":[4100, 4600, 4600, 5100, 4100, 5000, 5100],
    "할인율":[0.5, 0.1, 0.2, 0.3, 0, 0, 0],
    "칼로리":[10, 180, 420, 320, 20, 500, 400],
}
data = pd.DataFrame(data)
data.to_csv('data.csv', index=False)
```

```python
# 데이터 불러오기
df = pd.read_csv('data.csv')
df
```

# loc

```python
# 인덱싱 (행 전체)
df.loc[0]
```

```python
# 슬라이싱 (컬럼 전체)
df.loc[:,'가격']
```

```python
# 슬라이싱
df.loc[1,'가격']
```

```python
df.head()
```

```python
# 카페모카 메뉴와 가격
df.loc[2,'메뉴':'가격']
```

```python
# 카페모카 메뉴와 칼로리
df.loc[2, ['메뉴','칼로리']]
```

```python
# 카페라떼, 카페모카 메뉴와 가격
df.loc[1:2,'메뉴':'가격']
```

# iloc

```python
# 인덱싱 (행 전체)
df.iloc[0]
```

```python
# 슬라이싱 (컬럼 전체)
df.iloc[:, 1]
```

```python
# 카페모카 메뉴와 가격
df.iloc[2, 0:2]
```

```python
df.iloc[1:3]
```

# 데이터 추가

```python
# 결측값으로 추가, 원두 컬럼을 만들고 결측값(NaN)으로 대입
import numpy as np
df['원두'] = np.nan
df
```

```python
# 아메리카노의 원두는 '콜롬비아' 데이터 추가
df.loc[0,'원두'] = '콜롬비아'
df
```

```python
# 리스트 형태로 데이터 (행) 추가
df.loc['시즌'] = ['크리스마스라떼', 6000, 0, 500, '한국']
df
```

```python
# 딕셔너리 형태로 데이터(행) 추가
df.loc[7] = {'메뉴':'에소프레소', '가격':2000, '칼로리':10}
df
```

# 소팅

```python
# 데이터 보기
df
```

```python
df = df.drop('시즌', 0)
```

```python
# 인덱스 기준
df.sort_index(ascending = False)
```

```python
# 값 기준
# 메뉴 기준 정렬
df.sort_values('메뉴')
```

```python
# 가격과 메뉴 기준 정렬
df = df.sort_values(['가격','메뉴'],ascending = [False, True])
df
```

```python
# 인덱스 새로 만들기
df.reset_index(drop=True)
```

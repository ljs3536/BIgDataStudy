# 데이터 불러오기

# 데이터 전처리 & 피처엔지니어링

```python
# baseline (object type 컬럼 삭제)

cols = train.select_dtypes(include=object).columns

# print(train.shape, test.shape)
# train = train.drop(cols,axis=1)
# test = test.drop(cols,axis=1)
# print(train.shape, test.shape)

# label
from sklearn.preprocessing import LabelEncoder
for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
train[cols].head()

# one-hot
# train = pd.get_dummies(train, columns=cols)
# test = pd.get_dummies(test, columns=cols) 
```

```python
train = train.drop('CLIENTNUM',axis=1)
test_id = test.pop('CLIENTNUM')
```

```python
train.head(1)
```

```python
test_id
```

# 검증데이터 분리

```python
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train.drop('Attrition_Flag',axis=1), train['Attrition_Flag'], test_size=0.2, random_state=2022
                                           )
```

```python
X_tr.head()
```

# 모델 & 평가

```python
#모델불러오기
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=2022)
#핏(학습)
model.fit(X_tr, y_tr)
#예측
pred = model.predict(X_val)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

#정확도
print(accuracy_score(y_val, pred))

#정밀도
print(precision_score(y_val, pred))

#재현율 (민감도)
print(recall_score(y_val, pred))

#F1
print(f1_score(y_val, pred))

# baseline
# 0.9666872301048736
# 0.9444444444444444
# 0.8435114503816794
# 0.8911290322580645

# label
# 0.9685379395434917
# 0.948936170212766
# 0.851145038167939
# 0.89738430583501

# one-hot
# 0.9605181986428131
# 0.9419642857142857
# 0.8053435114503816
# 0.868312757201646
```

```python
pred_proba = model.predict_proba(X_val)
#roc-auc
print(roc_auc_score(y_val, pred_proba[:,1]))

# baseline
# 0.9894048160692921

# lable
# 0.9899707351049547

# one-hot
# 0.9885945548197205
```
